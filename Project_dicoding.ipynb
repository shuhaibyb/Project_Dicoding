### Cleaning

Untuk dibagian cleaning, saya bagi menjadi dua


*   column column yang berisi data numerik akan saya interpolasi dikarenakan dataset ini berisi data time series
*   untuk column wd karena data kategorikal akan saya isi dengan nilai dari diatasnya



```python
def interpolasi(datasets, columns):
  for dataset in datasets:
    for column in columns:
      dataset[column].interpolate(method='linear', limit_direction='forward', inplace=True)

def imputasi_wd(datasets):
  for dataset in datasets:
    dataset['wd'] = dataset['wd'].fillna(method='ffill')
```

```python
columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
datasets = [df_Aotizhongxin, df_Changping, df_Dingling, df_Dongsi, df_Guanyuan, df_Gucheng, df_Huairou, df_Nongzhanguan, df_Shunyi, df_Tiantan, df_Wanliu, df_Wanshouxigong]

interpolasi(datasets, columns)

imputasi_wd(datasets)
```

```python
x = 1

for dataset in datasets:
  print("dataset: ", x)
  print(dataset.isna().sum())
  print("\n")
  x = x + 1
```

### Merge Seluruh df & Pivot Pertanggal

```python
datasets = [df_Aotizhongxin, df_Changping, df_Dingling, df_Dongsi, df_Guanyuan, df_Gucheng, df_Huairou, df_Nongzhanguan, df_Shunyi, df_Tiantan, df_Wanliu, df_Wanshouxigong]

df_concat = pd.concat(datasets, axis=0)

df_concat
```

```python
df_concat.corr(numeric_only=True)
```

```python
df_concat.groupby(by = ['date', 'station']).agg({
    'PM2.5' : 'mean',
    'PM10' : 'mean',
    'SO2' : 'mean',
    'NO2' : 'mean',
    'CO' : 'mean',
    'O3' : 'mean',
    'TEMP' : 'mean',
    'PRES' : 'mean',
    'DEWP' : 'mean',
    'RAIN' : 'mean',

})
```

```python
df_concat['month_year'] = pd.to_datetime(df_concat['month'].astype('str') + '-' + df_concat['year'].astype('str'), format='%m-%Y')

df_concat.groupby(by = ['month_year']).agg({
    'PM2.5' : 'mean',
    'PM10' : 'mean',
    'SO2' : 'mean',
    'NO2' : 'mean',
    'CO' : 'mean',
    'O3' : 'mean',
    'TEMP' : 'mean',
    'PRES' : 'mean',
    'DEWP' : 'mean',
    'RAIN' : 'mean',

})
```

```python
df_concat['year'] = pd.to_datetime(df_concat['year'], format='%Y')

df_concat.info()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1 : Bagaimana Konsentrasi Polutan Udara (PM2.5, PM10, CO, SO2, NO2, dan O3) dari 2013 - 2017?

```python
df_pertanyaan1 = df_concat.groupby(by = ['year'], as_index = False).agg({
    'PM2.5' : 'mean',
    'PM10' : 'mean',
    'SO2' : 'mean',
    'NO2' : 'mean',
    'CO' : 'mean',
    'O3' : 'mean',
})
df_pertanyaan1
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_pertanyaan1['year'], df_pertanyaan1['PM2.5'], label='PM 2.5', color='red')
plt.plot(df_pertanyaan1['year'], df_pertanyaan1['PM10'], label='PM 10', color='orange')
plt.plot(df_pertanyaan1['year'], df_pertanyaan1['SO2'], label='SO2', color='magenta')
plt.plot(df_pertanyaan1['year'], df_pertanyaan1['NO2'], label='NO2', color='brown')
plt.plot(df_pertanyaan1['year'], df_pertanyaan1['O3'], label='O3', color='lightblue')
plt.title('Konsentrasi Polutan Udara', size=20)
plt.xlabel('Year',size=15)
plt.ylabel('Konsentrasi (μg/m³)',size=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_pertanyaan1['year'], df_pertanyaan1['CO'], label='CO', color='purple')
plt.title('Konsentrasi Polutan Udara', size=20)
plt.xlabel('Year',size=15)
plt.ylabel('Konsentrasi (μg/m³)',size=15)
plt.legend()
plt.show()
```

### Pertanyaan 2 : Bagaimana Perkembangan Curah Hujan per Bulannya?

```python
df_pertanyaan2 = df_concat.groupby(by = ['month_year'], as_index = False).agg({
    'RAIN': 'mean'
})
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_pertanyaan2['month_year'], df_pertanyaan2['RAIN'], label='Curah Hujan', color='blue')
plt.title('Curah Hujan', size=20)
plt.xlabel('Month Year',size=15)
plt.ylabel('mm/hari',size=15)
plt.legend()
plt.show()
```
